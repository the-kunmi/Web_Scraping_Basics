```python
from selenium.webdriver import Chrome
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

# Set up Chrome WebDriver
service = Service(executable_path=r"C:\Users\yaday\chromedriver-win64\chromedriver.exe")
driver = webdriver.Chrome(service=service)

# Open GitHub Trending page
driver.get("https://github.com/trending")
driver.implicitly_wait(10)

# Select "This month" filter
try:
    time_dropdown = WebDriverWait(driver, 15).until(EC.element_to_be_clickable((By.XPATH, "//details[@id='select-menu-date']")))
    time_dropdown.click()

    time.sleep(5)
    
    this_month_option = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//span[contains(text(), 'This month')]")))
    this_month_option.click()
    
    time.sleep(20)
    
    language_dropdown = WebDriverWait(driver, 15).until(EC.element_to_be_clickable((By.XPATH, "//details[@id='select-menu-language']")))
    language_dropdown.click()
   
    python_option = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//span[contains(text(), 'Python')]")))
    python_option.click()
   
except Exception as e:
    print(e)

time.sleep(40)

# Get list of all repositories
repos = driver.find_elements(By.XPATH, "//article[@class='Box-row']")

repo_names = []
repo_owners = []
stars = []
forks = []
descriptions = []
for repo in repos:
    try:
        # Extract repository name
        name_element = repo.find_element(By.XPATH, ".//h2[@class='h3 lh-condensed']")
        repo_name = name_element.text.strip()
        repo_names.append(repo_name)

        desc = repo.find_elements(By.CLASS_NAME, "col-9.color-fg-muted.my-1.pr-4")
        if desc:  # Check if the list is not empty
           descriptions.append(desc[0].text.strip())
        else:
           descriptions.append('None') # Append to the list

        star = repo.find_elements(By.XPATH, ".//a[contains(@href, '/stargazers')]")
        if star:
            stars.append(star[0].text.strip())
        else:
            stars.append('None')

        fork =  repo.find_elements(By.XPATH, ".//a[contains(@href, '/forks')]")
        if fork:
            forks.append(fork[0].text.strip())
        else:
            forks.append('None')
      
        # Extract contributor links (owners)
        contributors_section = repo.find_elements(By.XPATH, ".//span[contains(@class, 'd-inline-block')]/a")
        if contributors_section:
            owners = [contributor.get_attribute("href").split("/")[-1] for contributor in contributors_section]
        else:
            owners = []  # No contributors found

    except Exception as e:
        print(f"Error processing repo '{repo_name}': {e}")
        owners = []

    repo_owners.append(owners)

# Print extracted data
#for i in range(len(repo_names)):
    #print(f"Repo: {repo_names[i]}")
    #print(f"Owners: {repo_owners[i]}")
    #print("-" * 80)
    #print(descriptions)
   # print(stars)
    #print(forks)

df = pd.DataFrame({
    'Repository': repo_names,
    'Description': descriptions,
    'Stars': stars,
    'Forks': forks,
    'Built_by': repo_owners
})
# Close the browser
driver.quit()
df

```
